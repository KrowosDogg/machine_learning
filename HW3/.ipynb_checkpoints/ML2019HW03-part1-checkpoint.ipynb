> In the late 1950’s Charles Keeling invented an accurate way to measure atmospheric $CO_2$ concentration and began taking regular measurements at the Mauna Loa observatory.


Take `monthly_co2_mlo.csv` file, load it and prepare the data.

3. Normalize the target value by fitting a transformation on the train
4. Plot the resulting target against the time index

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, Normalizer
import matplotlib.pyplot  as plt

### BEGIN Solution
data = pd.read_csv('./data/monthly_co2_mlo.csv')
data = data.replace(-99.99, np.NaN).dropna().reset_index(drop = True)
x = data ['year'] - data['year'].min()
data = data.drop('year', axis = 1)
y = data ['CO2 [ppm]']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.40, random_state=32, shuffle = False)


#allin = pd.concat([X_train, y_train], axis = 1).sort_values('year')
#X_train = allin['year']
#y_train = allin['CO2 [ppm]']

#scaler = StandardScaler()

#allout = pd.concat([X_test, y_test], axis = 1).sort_values('year')
#X_test = allout['year']
#y_test = allout['CO2 [ppm]']    

#y_train = scaler.fit_transform(y_train.values.reshape(-1, 1))
#y_test = scaler.transform(y_test.values.reshape(-1, 1))

plt.figure(figsize = (15, 5))
plt.scatter(X_train, y_train, alpha = 0.4, s = 10)
plt.plot(X_train, y_train)
plt.show()
### END Solution
````

### Task 4.2 (2 pt.)

Use [**GPy**](https://pypi.python.org/pypi/GPy) library for training and prediction. Fit a GP and run the predict on the test. Useful kernels to combine: `GPy.kern.RBF, GPy.kern.Poly, GPy.kern.StdPeriodic, GPy.kern.White, GPy.kern.Linear`. 

1. Plot mean and confidence interval of the prediction. 
2. Inspect them on normality by scatter plot: plot predicted points/time series against true values. 
3. Estimate the prediction error with `r2_score`. R2-score accepted > 0.83 on test sample.

````python
from GPy.models import GPRegression
import GPy
from GPy.kern import RBF, Poly, StdPeriodic, White, Linear
from sklearn.metrics import r2_score
````

````python
kernel = GPy.kern.RBF(1, lengthscale=0.1) + GPy.kern.Poly(1)
model = GPy.models.GPRegression(X_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1), kernel)
model.optimize()
model
````

````python
#X_t = scaler.transform(X_test.values.reshape(-1, 1))
prediction, variance = model.predict(X_test.values.reshape(-1, 1))
std = np.sqrt(variance)
plt.figure(figsize = (15, 5))
#plt.plot(model.X, model.Y, 'kx', label='Train data')

plt.plot(X_test, prediction, 'kx',label='Data')
plt.plot(X_test, prediction, label='Prediction')
plt.fill_between(X_test, (prediction - std).reshape(291,), (prediction + std).reshape(291,), alpha=0.3)
plt.legend()    
plt.xlabel('Year', fontsize=14)
plt.ylabel('Conc', fontsize=14)
````

````python
plt.figure(figsize=(10, 10))
plt.scatter(y_test, prediction)
#plt.plot(y_test, prediction)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
````

````python
r2_score(y_test, prediction)
````
